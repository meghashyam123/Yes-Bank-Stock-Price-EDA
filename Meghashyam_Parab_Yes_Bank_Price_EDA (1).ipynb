````python

#create a function to find outliers using IQR

def find_outliers_IQR(df):
   q1=df.quantile(0.25)
   q3=df.quantile(0.75)
   IQR=q3-q1
   outliers = df[((df<(q1-1.5*IQR)) | (df>(q3+1.5*IQR)))]
   return outliers
````

````python

# find outliers in percentage
for i in ['Open' , 'High', 'Low', 'Close']:
  outliers = find_outliers_IQR(stock_df[i])
  outliers_percent = round((outliers.size/stock_df[i].size)*100, 2)
  print(f'{i} : {outliers_percent}%')
````

````python
# find outliers
for i in ['Open' , 'High', 'Low', 'Close']:
  outliers = find_outliers_IQR(stock_df[i])
  print(f'{i} : ')
  print(outliers)
````

#### Chart - 5

````python
# Chart - 5 visualization code

import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a date column, derive the quarter and year
stock_df['Quarter'] = stock_df.index.to_period('Q')
stock_df['Year'] = stock_df.index.year

# Filter data for the last 4 years
last_4_years = stock_df['Year'].unique()[-4:]  # Get the last 4 years
filtered_df = stock_df[stock_df['Year'].isin(last_4_years)]

# Group by quarter and aggregate
quarterly_data = filtered_df.groupby(['Year', 'Quarter']).aggregate(
    {'Open': 'mean', 'High': 'mean', 'Low': 'mean', 'Close': 'mean'}
)

# Plotting the data for each quarter in the last 4 years
fig, ax = plt.subplots(figsize=(16, 8))  # Adjust figsize for better visualization

width = 0.15  # Width of each bar
x_pos = range(len(quarterly_data.index))

ax.bar([x - width * 1.5 for x in x_pos], quarterly_data['Open'], width, label='Open')
ax.bar([x - width * 0.5 for x in x_pos], quarterly_data['High'], width, label='High')
ax.bar([x + width * 0.5 for x in x_pos], quarterly_data['Low'], width, label='Low')
ax.bar([x + width * 1.5 for x in x_pos], quarterly_data['Close'], width, label='Close')

ax.set_xticks(x_pos)
ax.set_xticklabels([f'{year}Q{quarter.quarter}' for year, quarter in quarterly_data.index])  # Format labels
ax.set_ylabel('Average stock price')
ax.set_title('Average Stock Prices by Quarter (Last 4 Years)')
ax.legend()
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for readability

plt.show()
````

Chart - 11

````python
# Define the Bollinger Bands window (typically 20 days)
bollinger_window = 20

# Compute the Simple Moving Average (SMA)
stock_df["SMA"] = stock_df["Close"].rolling(window=bollinger_window).mean()

# Compute the Standard Deviation
stock_df["Rolling Std"] = stock_df["Close"].rolling(window=bollinger_window).std()

# Compute the Upper and Lower Bollinger Bands
stock_df["Upper Band"] = stock_df["SMA"] + (2 * stock_df["Rolling Std"])
stock_df["Lower Band"] = stock_df["SMA"] - (2 * stock_df["Rolling Std"])

# Plot the Closing Price and Bollinger Bands
plt.figure(figsize=(12, 6))
# If 'Date' is in the index, use stock_df.index:
plt.plot(stock_df.index, stock_df["Close"], color="blue", label="Close Price")
plt.plot(stock_df.index, stock_df["SMA"], color="orange", label="SMA (20-day)")
plt.plot(stock_df.index, stock_df["Upper Band"], color="red", linestyle="--", label="Upper Band")
plt.plot(stock_df.index, stock_df["Lower Band"], color="green", linestyle="--", label="Lower Band")

# Labels and title
plt.xlabel("Date")
plt.ylabel("Stock Price")
plt.title("Yes Bank Bollinger Bands (20-Day)")
plt.legend()
plt.grid()

# Show the plot
plt.show()
````

##### 1. Why did you pick the specific chart?

Bollinger Bands were chosen because they provide a powerful visual representation of stock volatility, trend direction, and potential buy/sell signals.

‚úÖ Tracks Market Volatility ‚ö°

The bands expand when volatility is high and contract when it‚Äôs low, helping traders anticipate market movements.

‚úÖ Identifies Trading Signals üîç

Price near the upper band ‚Üí Overbought (sell signal)
Price near the lower band ‚Üí Oversold (buy signal)

‚úÖ Detects Trend Reversals üîÑ

A stock consistently touching the upper band in an uptrend can signal strong bullish momentum.
A breakdown below the lower band (like in 2018-2020 for Yes Bank) warns of potential crashes.
